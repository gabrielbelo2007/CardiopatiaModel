### Importações

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
```

### Treinamento

```python
objetivo = 'BP_Category'

df_completo = pd.read_csv(url_features, sep=",")
df_completo = df_completo.drop('HighRisk', axis=1)

print(df_completo[objetivo].value_counts())

colunas = df_completo.select_dtypes(include=['object']).columns
for coluna in colunas:
    if coluna in df_completo.columns:
        le = LabelEncoder()
        df_completo[coluna] = le.fit_transform(df_completo[coluna].astype(str))

x = df_completo.drop(objetivo, axis=1)
y = df_completo[objetivo]

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
x_treino = scaler.fit_transform(x_treino)
x_teste = scaler.transform(x_teste)

knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(x_treino, y_treino)

y_resultado = knn.predict(x_teste)
print("\nRelatório de Classificação:\n", classification_report(y_teste, y_resultado))
```
